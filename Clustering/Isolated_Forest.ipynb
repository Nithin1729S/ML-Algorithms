```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.ensemble import IsolationForest
```

```python
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
```

```python
np.random.seed(42)
n_outliers = 20
outliers = np.random.uniform(low=-10, high=10, size=(n_outliers, 2))
X_with_outliers = np.vstack([X, outliers])
```

```python
plt.scatter(X_with_outliers[:,0],X_with_outliers[:,1])
plt.show()
```

```python
clf=IsolationForest(contamination=0.1)
clf.fit(X_with_outliers)
```

```python
predictions=clf.predict(X_with_outliers)
```

```python
predictions
```

```python
plt.scatter(X_with_outliers[:,0],X_with_outliers[:,1],c=predictions)
plt.show()
```
